```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#creating dataframe from array
data = np.random.randint(low=1, high=100, size=(100,4))
df = pd.DataFrame(data, columns=list('ABCD'))
print(df)
#df.head()
```

```python
#cumulative option
df.hist(bins=10)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
df.hist(bins=30, cumulative=True)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#normal distribution using numpy
mu, sigma = 0, 0.1 #mean and standard deviation
d = np.random.normal(mu, sigma, 10000)
plt.hist(d, 100, density=True)
```

```python
#normal distribution using seaborn
from scipy.stats import norm
# generate random numbers from N(0,1)
data_normal = norm.rvs(size=10000,loc=0,scale=1)
ax = sb.distplot(data_normal,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Normal Distribution', ylabel='Frequency')
```

```python
#make more informative using Seaborn, labelling x And y
import seaborn as sb
sb.distplot(df['A'], kde=False, label='A')
sb.distplot(df['B'], kde=False, label='B')

# Plot formatting
plt.legend(prop={'size': 12})
plt.title('Number Distribution')
plt.xlabel('Number')
plt.ylabel('Frequency')  
plt.show()
```

```python
#Transform to inverse logit
from scipy.special import logit, expit
x = np.random.normal(mu, sigma, 10000)
y = logit(x)
plt.plot(x, y)
plt.grid()
plt.ylim(-6, 0)
plt.xlabel('x')
plt.title('logit(x)')
plt.show()
```

```python
#poisson distribution using numpy
s = np.random.poisson(3, 10000) #lamda, size
count, bins, ignored = plt.hist(s, 20, density=False)
plt.show()
#how mean and standard deviation influence the shape of histogram and cdf?
```

```python
#poisson distribution using seaborn
from scipy.stats import poisson
data_poisson = poisson.rvs(mu=3, size=10000) #mu/lamda
ax = sb.distplot(data_poisson,
                  bins=20,
                  kde=False,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Poisson Distribution', ylabel='Frequency')
#mean?
#variance?
```

```python
#beta distribution
from scipy.stats import beta
a, b = 2.30984964515, 0.62687954301
r = beta.rvs(a, b, size=10000)
plt.hist(r, density=True)
plt.show()
#location dispersion parametrization?
#mean-population size?
```
